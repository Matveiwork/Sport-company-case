```python
import pandas as pd
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
import numpy as np
from efficient_apriori import apriori
```

```python
df = pd.read_pickle(' all_data.pkl ')
```

```python
df
```

```python
df['buy'] = None
```

```python
dfno = df.loc[df['dt']==0]
dfno
```

```python
dfye = df.loc[df['dt']>0]
```

```python
dfye['buy'] = dfye['buy'].apply(lambda x: 'yes')
```

```python
dfno['buy'] = dfno['buy'].apply(lambda x: 'no')
```

```python
dfans = pd.concat([dfye,dfno])
```

```python
dfans.head()
```

```python
df_categorial = dfans[['education','product','buy']]
```

```python
columns = list(df_categorial.columns)

transactions = []
for i in range(len(df_categorial)):
    values=tuple(list(t.strip() for t in df_categorial.iloc[i,:].values if (t is not None) and (not isinstance(t,float))))
    transactions.append(values)
```

```python
items,rules = apriori(transactions,min_support=0.003,min_confidence=0.4)
```

```python
rules
```

```python
with open('ids_first_company_negative.csv', 'r') as file:
     content = file.read()
```

```python
import re
cleaned_text3 = re.sub(r'[a-zA-Zа-яА-Я]', '', content)
```

```python
delimiters = (",", ";", "\n", " ")
```

```python
result3 = re.split('|'.join(map(re.escape, delimiters)), cleaned_text3)
```

```python
df7 = pd.DataFrame(result3[5:-2])

df7 = df7.rename(columns={0:'id'})[['id']]

df7['id'] = pd.to_numeric(df7['id'], errors='coerce')


df7 = df7.dropna(subset=['id'])

print(df7)
```

```python
value_to_remove = 0

df7 = df7[(df7['id'] != value_to_remove)]
```

```python
df_negative = df7
df_negative
```

```python
df_negative['id']=df_negative['id'].astype(int)
```

```python
df_positive['id']=df_positive['id'].astype(int)
```

```python
df = pd.read_pickle(' all_data.pkl ')
```

```python
df
```

```python
df_com1 = df[(df['dt'] >= 5) & (df['dt'] <= 16)]
```

```python
df_com1
```

```python
negative = df_negative.to_dict()
positive = df_positive.to_dict()
```

```python
df_com1_pos = pd.merge(df_com1,df_positive,left_on = 'id',right_on = 'id',how='inner')
```

```python
df_com1_neg = pd.merge(df_com1,df_negative,left_on = 'id',right_on = 'id',how='inner')
```

```python
df_com1_pos
```

```python
df_com1_neg
```

```python
test_cash = df_com1_pos['cost'].to_list()
control_cash = df_com1_neg['cost'].to_list()
```

```python
test = np.array(test_cash)
control = np.array(control_cash)
```

```python
print(ttest_ind(test_cash,control_cash,equal_var=False))
print('средняя выручка тестовой группы=',test.mean())
print('средняя выручка контрольной группы=',control.mean())
```

```python
print('общая выручка тестовой группы=',test.sum())
print('общая выручка контрольной группы=',control.sum())
```
